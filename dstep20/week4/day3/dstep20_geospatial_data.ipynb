```
# -- let's read a file for example
gdat = gp.read_file("https://data.cityofnewyork.us/api/geospatial/6j86-5s7z?method=export&format=GeoJSON")
```

```
# -- print the data to screen
gdat
```

Geospatial data file formats come in a couple of different flavors.  The most common is called a "shapefile" which actually consists of several files:

- **filename.shp** – the geometries of the shapes
- **filename.shx** – the index of the shapes 
- **filename.dbf** – the features associated with the shapes
- **filename.prj** – (optional) the coordinate system or "projection"

### **Providing Spatial Context for Datasets**

Let's go back to our heatmap calendar example for a moment.


```
# -- import pandas, numpy, and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# -- set the filename and read the data while parsing the MONTH column
fname = "https://data.cityofnewyork.us/api/views/ebb7-mvp5/rows.csv?accessType=DOWNLOAD"
dsny  = pd.read_csv(fname, parse_dates=["MONTH"])

# -- sub-select only the first six columns
cols = dsny.columns[[0, 1, 2, 3, 4, 5, 10]]
dsny = dsny[cols]

# -- rename the columns for ease of use
dsny.columns = ["month", "borough", "district", "refuse", "paper", "mgp", "borough_id"]
```

```
# -- print the columns of both the geospatial and non-geospatial data sets
pd.set_option('display.max_columns', 500)
print(gdat.head())
print("")
print(dsny.head())
```

Notice that these two data sets actually do have some commonality: districts.

```
# -- print district features of the data sets
print(gdat["districtcode"])
print(dsny["district"])
```

```
# -- create a column in tonnage data that corresponds to column in geospatial data
dsny["dcode"] = dsny["borough_id"].astype(str) + dsny["district"].astype(str).str.zfill(2)
```

```
# -- print it out
dsny["dcode"]
```

```
print(gdat.head())
print("")
print(dsny.head())
```

```
dsny["dcode"].iloc[0]
```

```
# -- now let's get the total tonnage for each of those districts by grouping by
#    district and summing (also resetting index)
nyc_dtot = dsny.groupby("dcode").sum().reset_index()
```

```
nyc_dtot
```

Now that we have two columns that "align", we can "merge" (or "join") the GeoDataFrame and DataFrame:

```
# -- merge the two (nb, the DataFrame must be merged into the GeoDataFrame)
gdatm = gdat.merge(nyc_dtot, left_on="districtcode", right_on="dcode")
```
